```python
# Import libraries
import pandas as pd
import numpy as np
import urllib.request
```

# Data Preparation

```python
# Download and load data
data_url = "https://raw.githubusercontent.com/alexeygrigorev/mlbookcamp-code/refs/heads/master/chapter-02-car-price/data.csv"
urllib.request.urlretrieve(data_url, "data.csv")

data = pd.read_csv("data.csv")
data.head()
```

```python
data.columns = data.columns.str.replace(" ", "_").str.lower()
data.dtypes
```

```python
strings = [*data.dtypes[data.dtypes == 'object'].index]
```

```python
for col in strings:
    data[col] = data[col].str.lower().str.replace(" ", "_")
```
